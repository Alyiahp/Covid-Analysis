```python
import pandas as pd
pd.options.plotting.backend = "plotly"
import plotly.graph_objects as go
from matplotlib import pyplot as plt
import plotly.express as px
import numpy as np
from IPython.display import Image
```

Reading in the needed data

```python
cases = pd.read_csv("../../../../data/stage_2/covid_confirmed_usafacts.csv")
deaths =  pd.read_csv("../../../../data/stage_2/covid_deaths_usafacts.csv")
population_perCounty = pd.read_csv("../../../../data/stage_2/covid_county_population_usafacts.csv")
```

I will be choosing Arizona for my state analysis, so I will need to only get the columns that contain the state name AZ

```python
deaths = deaths.drop(deaths.index[cases['countyFIPS'] == 0])
cases = cases.drop(cases.index[cases['countyFIPS'] == 0])

cases = cases.groupby('State')
chunks = dict(list(cases))
AZ_cases = chunks['AZ']

deaths = deaths.groupby('State')
chunks = dict(list(deaths))
AZ_deaths = chunks['AZ']
```

```python
AZ_cases.head()
```

```python
AZ_deaths.head()
```

Calculating new deaths and cases by subtracting the day before from the day after. 

```python
AZ_new_deaths = pd.DataFrame()

for column in range(5,577):
  AZ_new_deaths[column] = AZ_deaths.iloc[:,column] - AZ_deaths.iloc[:,column-1] 


AZ_new_cases = pd.DataFrame()

for column in range(5,577):
  AZ_new_cases[column] = AZ_cases.iloc[:,column] - AZ_cases.iloc[:,column-1] 
```

Generating the weekly statistics by grouping the sum of 7 days 

```python
AZ_week_deaths = pd.DataFrame()
AZ_week_cases = pd.DataFrame()

AZ_week_deaths = AZ_new_deaths.groupby([[i//7 for i in range(0, 572)]], axis = 1).sum()
AZ_week_cases = AZ_new_cases.groupby([[i//7 for i in range(0, 572)]], axis = 1).sum()
```

Getting the average cases and deaths for the entire state 

```python
#Average cases
AZ_avg_cases = pd.DataFrame
AZ_avg_cases= pd.DataFrame(AZ_week_cases.mean(axis=0).round())
AZ_avg_cases.columns = ['AZ_cases_avg']

#Average deaths
AZ_avg_deaths = pd.DataFrame
AZ_avg_deaths= pd.DataFrame(AZ_week_deaths.mean(axis=0).round())
AZ_avg_deaths.columns = ['AZ_deaths_avg']

#joining them 
AZ_averages = AZ_avg_cases.join(AZ_avg_deaths)
AZ_averages
```

Generating basic statistics for Arizona new cases and deaths 

```python
AZ_averages.mean(axis=0)
```

```python
AZ_averages.mode(axis=0)
```

```python
AZ_averages.median(axis=0)
```

We will now compare the top 5 infected counties in Arizona. 

```python
#getting county names 
counties = AZ_cases['County Name']. drop_duplicates()

#getting Adding county names to weekly sum data calulated earlier
AZ_deaths_counties = pd.DataFrame()
AZ_cases_counties = pd.DataFrame()

#Getting sum of all cases
AZ_cases_counties['total_cases'] = AZ_cases.sum(axis=1)
AZ_deaths_counties['total deaths'] = AZ_deaths.sum(axis=1)

#joining the sum and county names 
AZ_cases_counties= pd.DataFrame(AZ_cases_counties.join(counties))
AZ_deaths_counties =pd.DataFrame( AZ_deaths_counties.join(counties))
```

```python
## sorting the counties by cases
AZ_cases_counties = AZ_cases_counties.sort_values(by=['total_cases'],ascending=False)
AZ_cases_counties
```

Now lets compare the counties using normalized log values. 

```python
AZ_week_deaths_log = AZ_week_deaths.loc(axis=0)[106,109,110,113,107]
AZ_week_cases_log = AZ_week_cases.loc(axis=0)[106,109,110,113,107]

#Normalizing deaths
AZ_week_deaths_log.loc(axis=0)[106] = AZ_week_deaths_log.loc(axis=0)[106].apply(lambda x: x/(np.log(4734447)))
AZ_week_deaths_log.loc(axis=0)[109] = AZ_week_deaths_log.loc(axis=0)[109].apply(lambda x: x/(np.log(1079454)))
AZ_week_deaths_log.loc(axis=0)[110] = AZ_week_deaths_log.loc(axis=0)[110].apply(lambda x: x/(np.log(510738)))
AZ_week_deaths_log.loc(axis=0)[113] = AZ_week_deaths_log.loc(axis=0)[113].apply(lambda x: x/(np.log(220312)))
AZ_week_deaths_log.loc(axis=0)[107] = AZ_week_deaths_log.loc(axis=0)[107].apply(lambda x: x/(np.log(220848)))
```

```python
#Normalizing cases
AZ_week_cases_log.loc(axis=0)[106] = AZ_week_cases_log.loc(axis=0)[106].apply(lambda x: x/(np.log(4734447)))
AZ_week_cases_log.loc(axis=0)[109] = AZ_week_cases_log.loc(axis=0)[109].apply(lambda x: x/(np.log(1079454)))
AZ_week_cases_log.loc(axis=0)[110] = AZ_week_cases_log.loc(axis=0)[110].apply(lambda x: x/(np.log(510738)))
AZ_week_cases_log.loc(axis=0)[113] = AZ_week_cases_log.loc(axis=0)[113].apply(lambda x: x/(np.log(220312)))
AZ_week_cases_log.loc(axis=0)[107] = AZ_week_cases_log.loc(axis=0)[107].apply(lambda x: x/(np.log(220848)))

#joining the table with counties 
AZ_week_deaths_log = AZ_week_deaths_log.join(counties)
AZ_week_cases_log = AZ_week_cases_log.join(counties)
```

Plotting

```python
AZ_week_deaths_log = AZ_week_deaths_log.set_index('County Name')
fig = AZ_week_deaths_log.T.plot(title='County Comparisons New deaths')
fig.update_layout(hovermode="x unified")

img_bytes = fig.to_image(format="png")
Image(img_bytes)
```

As you can see in both plots the trends in cases and deaths remain the same even after normalizing the values, the scale of the cases changed however and are now easier to see. Maricopa County still has a significant more amount of cases and deaths than other counties. And the trends from the State peaks are also seen within the individual counties.
